```python
roles_usuarios = ["abministrador","analista","gerente","auditor"]
# ordenada : por orden de ingreso
print(roles_usuarios[3])
for i in roles_usuarios:
    print(roles_usuarios)
""""""
import random
nombres =["Toby","Gruñon","Beky","Chiquitin","Pacho","Garrapata"]

nombre_secreto = random.choice(nombres)

print("==== ADIVINA LA MASCOTA ====")
print("Tiene 3 intentos de vida..")
vidas = 3
while vidas > 0:
    for nombre in nombres:
        print(nombre)
    intento = input("Ingrese el nombre de la Mascota:")
    if intento.lower() == nombre_secreto.lower():
        print("Mascota ganaste,adivinaste el nombre..")
        break
    else:
        vidas  -=1
        print("ERROR CRITICO,el nombre de la Mascota no es.Intentos restantes:",vidas)
if vidas == 0:
    print("PAILAS,PERDISTE EL NOMBRE ERA:",nombre_secreto,"ahora consigna a mi nequi...")
""""""
#Ejercicio de gestion de clientes en una tienda
#Agregar nuevos clientes
#Recorrer la lista y mostrar todos los datos
#Modificar un nombre en caso de error
#Eliminar un cliente

#Funciones que utilizaremos
#len() -> cuenta las instancias dentro de la lista
#append() Añade datos a la lista
#title() coloca el primer caracter en mayuscula ej,carlos - Carlos
#pop() Elimina un dato de la lista


#Agregar nuevas mascotas
def agregar_mascotas(lista_mascotas,nombre):
    #Verificamos si el dato ingrsado sea una cadena y longitud valida entre 2 a 50 caracteres
    if isinstance(nombre,str) and 2 <= len(nombre) <=50:
        # si el nombre cumple
        lista_mascotas.append(nombre.title())
        print(f"Mascota agregada: {nombre.title()}")
    else:
        print("Nombre del cliente invalido debe tener entre 2 a 50 caracteres")
def mostrar_mascotas(lista_mascotas):
    for mascotas in lista_mascotas:
        print(f"-{mascotas}")

#Modificar un nombre en caso de error
def modificar_mascotas(lista_mascotas, indice,nuevo_nombre):
    #Verificamos si el dato ingrsado sea una cadena y longitud valida entre 2 a 50 caracteres
    if not isinstance(nuevo_nombre,str) and 2 <= len(nuevo_nombre) <=50:
        print("Nombre de  las Mascota invalido debe tener entre 2 a 50 caracteres")
        return
    if 0 <= indice <=len(lista_mascotas):
        original = lista_mascotas[indice]
        lista_mascotas[indice] = nuevo_nombre.title()
        print(f"La Mascota {original} fue modificado por:{nuevo_nombre.title()}")
    else:
        print("No existe ese cliente en la lista")
#Eliminar un cliente
#vamos a eliminar un cliente de la lista por indice
def eliminar_mascotas(lista_mascotas,indice):
    if 0 <= indice <=len(lista_mascotas):
        eliminado = lista_mascotas.pop(indice) #Eliminar el cliente de acuerdo al numero de indice
        #mostraremos el registro eliminado
        print(f"La Mascota eliminado: {eliminado}")
    else:
        print("Ese cliente no existe en nuestro sistema")


def main():
    mascota =["Toby","Gruñon","Beky","Chiquitin","Pacho","Garrapata"]
    print("Mascotas activos")
    mostrar_mascotas(mascota)
    agregar_mascotas(mascota,"Tony,Mechas,Rapumsel")
    print ("Mascotas activos mas el nuevo")
    mostrar_mascotas(mascota)
    modificar_mascotas(mascota,4,"Gabriela")
    mostrar_mascotas(mascota)
    eliminar_mascotas(mascota,1,4)
    mostrar_mascotas(mascota)

if __name__ =="__main__":
    main()
```
