```python
from abc import ABC, abstractmethod
from typing import Optional

# 1 interfaz
class Movible(ABC):
    #REPRESENTA animales que puedan desplazarce por si solo
    @abstractmethod
    def mover(self) ->None:
        pass
class Animal(ABC):
    def __init__(self,nombre:str)->None:
        self.__nombre:str ="" #privado y encapsulado
        self.nombre = nombre #Ya nombre puedo utilizarlo nombre(getter,setter)
#GETTER - MOSTRAR
@property
def nombre(self) ->None:
    return self.__nombre
@nombre.setter
def nombre(self,nuevo_nombre:str)->None:
  if isinstance(nuevo_nombre,str)and nuevo_nombre.strip()and nuevo_nombre !="":
    self.__nombre = nuevo_nombre.strip().title()
  else:
     raise ValueError("El nombre ingresado debe ser de una cadena de texto valida")
@abstractmethod
def sonido(self)->None:
   pass
#creremos clases hijas para las dos ABC
class Perro(Animal):
   def sonido(self)->None:
      print(f"{self.nombre}El perro hace Guau Guau")
```
