```python
class Libro: #Aca se crea la clase
    def __init__(self,titulo, precio): #Construct
        self.__titulo = titulo #Atributos, __ que es privado
        self.__precio = precio
# Constructor de metodos
   # getter, devuelve el valor del atributo solicitado. get

    def get_titulo(self): # Identifica el valor y solo lo muestra
       return self.__titulo
    def get_precio(self):  # Identifica el valor y solo lo muestra
       return self.__precio
  # Ahora realizamos un metodo que me muestre los dos valores
    def mostrar_info(self):
        print(f"\nEl titulo es: {self.__titulo},  el precio es: {self.__precio}\n")

  # Vamos a crear un metodo para cambiar el precio

  # Utilizamos SETTER (set), para realizar el cambio de el valor contenido en el
  # Atributo
    def set_precio(self, nuevo_precio):
        if isinstance(nuevo_precio, (int, float)) and nuevo_precio > 0:
             self.__precio = nuevo_precio
        else:
          print("Recuerde que debe ingresa un numero valido")
        #En el siguiente metodo vamos a cambiar el valor del atributo utilizando set
    def set_titulo(self, nuevo_titulo):
        if isinstance(nuevo_titulo, str) and len(nuevo_titulo) > 0: # Debe de ser string y tener al menos un let
           self.__titulo = nuevo_titulo # Actualiza el atributo con el nuevo valor
def main():
  # Aqui invoco los mètodos
  print("=======ESTIMADO APRENDIZ BIENVENIDO AL PROGRAMA DE LIBRO====\n")

  # Vamos a crear 2 objetos libro, uno es HTML, precio 200000
  # el segundo libro sera JAVASCRID. precio 186000
  libro1 = Libro("HTML", 200000)
  libro2 = Libro("JAVASCRID", 186000)

# Ahora vamos a mostrar los valores iniciales del libro1, con metodos separados
  print("\nINFORMACION DEL PRIMER LIBRO METODO SEPARADOS")
  print(f"El titulo es: {libro1.get_titulo()}")
  print(f"El precio es: {libro1.get_precio()}")

# Vamos a mostrar lo anterior pero con todo el metodo mostrar
  print("\nINFORMACION DEL PRIMER LIBRO UN METODO ")
  libro1.mostrar_info()

  # Ahora vamos actualizar el precio del libro
  print("\nATUALIZACION DEL PRECIO DEL PRIMER LIBRO")
  nuevo_precio = float(input("Ingrese el nuevo precio: "))
  libro1.set_precio(nuevo_precio)

  print("El nuevo precio actualizado del libro es: ")
  libro1.mostrar_info()

  # Ahora actualizaremos el nombre del libro
  print("\nACTUALIZACION DEL NOMBRE DEL PRIMER LIBRO")
  nuevo_titulo = input("Ingrese el nuevo titulo: ")
  libro1.set_titulo(nuevo_titulo)
  print("\nLos datos actualizados del libro son: ")
  libro1.mostrar_info()

if __name__ == "__main__":
  main()
```
